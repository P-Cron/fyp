## Summary
That is all the useful dynamic responses form the 08 Accord

* RPM
    * take once engine has heated up, past 30 degrees
    * Take min after that
    * Varies slightly from 819 to 821
* EGR Percent
    * Take while engine is idling
    * take bottom quantile
* Volumetric Efficiency
    * Not especially reliable
    * When engine not started (RPM at 0) then seems stable, at 58 or 59
    * do a non zero check first
